```python
# 7개 데이터프레임 병합
lr = pd.concat([lr2018, lr2019, lr2020, lr2021, lr2022, lr2023, lr2024], ignore_index=True)

lr
```

```python
# 행 이름 수정
lr = lr.rename(columns={'시군구': '주소'})

# 계약기간이 잘못 기입된 행 수정
lr['계약기간'] = lr['계약기간'].replace('022403~202603', '202403~202603')
lr['계약기간'] = lr['계약기간'].replace('202407~262607', '202407~202607')
lr['계약기간'] = lr['계약기간'].replace('202312~232603', '202312~202603')
lr['계약기간'] = lr['계약기간'].replace('202112~212312', '202112~202312')
lr['계약기간'] = lr['계약기간'].replace('201911~212111', '201911~202111')
lr['계약기간'] = lr['계약기간'].replace('201910~212110', '201910~202110')

# '계약기간'과 '계약구분'의 -(결측값)을 결측치로 대체
lr['계약기간'] = lr['계약기간'].replace('-', pd.NA)
lr['계약구분'] = lr['계약구분'].replace('-', pd.NA)

# '갱신요구권 사용' 열을 bool 타입으로 변환
lr['갱신요구권 사용'] = lr['갱신요구권 사용'].apply(lambda x: True if x == '사용' else False)

# '거래일자' 열 생성
lr['거래일자'] = pd.to_datetime(lr['계약년월'].astype(str) + lr['계약일'].astype(str), format='%Y%m%d')

# '시도', '시군구', '읍면동' 열 생성
lr[['시도', '시군구', '읍면동']] = lr['주소'].str.split(' ', expand=True)

# '번지'와 '도로명' 열 생성
lr[['계약일', '계약종료일']] = lr['계약기간'].str.split('~', expand=True)

# '계약일'과 '계약종료일'의 일자를 01로 대체
lr['계약일'] = lr['계약일'].apply(lambda x: x + '01' if pd.notnull(x) else x)
lr['계약종료일'] = lr['계약종료일'].apply(lambda x: x + '01' if pd.notnull(x) else x)

# '계약일'과 '계약종료일'을 datetime 타입으로 변환
lr['계약일'] = pd.to_datetime(lr['계약일'], format='%Y%m%d')
lr['계약종료일'] = pd.to_datetime(lr['계약종료일'], format='%Y%m%d')

# '계약 기간' 열 생성
lr['계약 기간(년)'] = (lr['계약종료일'] - lr['계약일']).dt.days // 365

# 행 순서 수정
columns = ['거래일자', '계약일', '계약종료일', '계약 기간(년)', '주소', '시도', '시군구', '읍면동', '번지', '도로명'] + [col for col in lr.columns if col not in ['거래일자', '계약일', '계약종료일', '계약 기간(년)', '주소', '시도', '시군구', '읍면동', '번지', '도로명']]
lr = lr[columns]

# 필요 없는 행 삭제
lr = lr.drop(columns=['계약년월', '본번', '부번', '주택유형', '계약기간'])

lr
```

```python
lr.to_csv('아파트(전월세)_실거래가_전처리완료.csv', index=False)
```
